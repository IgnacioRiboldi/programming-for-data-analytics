# Assigment 03 - Pie chart

#### The following program will create a pie chart from a CSV data source

##### By: Ignacio Riboldi

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download"
df = pd.read_csv(url)
count = df['Sex'].value_counts()
plt.pie(count, labels=count.index, colors=['Blue','Pink'], autopct='%1.1f%%')
plt.title('Distribution by Gender')
plt.show()
```
