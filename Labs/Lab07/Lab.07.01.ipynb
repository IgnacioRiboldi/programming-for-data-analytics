```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("https://cli.fusio.net/cli/climate_data/webdata/mly4935.csv", skiprows=19)

print(df.head(3))

# Checking correlations
corrtemp = df['month'].corr(df['meant'])

print(corrtemp)

```

```python

cleandf = df[["month","wdsp"]] # Cleaning data set. Dropping rows that are missing wind speed data

cleandf['wdsp']= cleandf.loc[:,('wdsp')].replace(' ', np.nan) # Converting 'NaN' strings to actual NaN values
cleandf.dropna(inplace=True) # Dropping rows with NaN values
```

```python
cleandf['wdsp'] = cleandf['wdsp'].astype(float) # Converting wind speed column to float
```

```python
sns.set_style('whitegrid')
#sns.scatterplot(x='total_bill', y='tip', data=dataset)
sns.lmplot(x='month', y='wdsp',order = 3, data=cleandf)
plt.show()
```
